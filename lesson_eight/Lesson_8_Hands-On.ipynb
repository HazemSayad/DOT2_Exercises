### The cell below is my favorite because I spent extra time on it to make it almost perfect and the results show!


1. **First**

    For `antiq = str(bisons.Species.value_counts()).split()`: the Species column gets a *`value_count()`*, then it is converted to a string using *`str()`*, then it is split using *`split()`* which creates an array that is then assigned to the variable **antiq**


2. **Second**

    The **antiq** variable was enumerated using *`enumerate()`* and each index was assigned to **i** for each value **v** in a for loop
    

3. **Third**

    For each iteration, an *if* condition was set to compare the value of *v* with the search term *antiquus*


4. **Finaly**

    When the *if* condition was met, the **antiq** variable was substituted for a single value that comes immediately after the word *antiq* (hence the index i+1), which would be the value for *antiquus* 

### Then the result was printed

```python
antiq = str(bisons.Species.value_counts()).split()
for i, v in enumerate(antiq):
    if v.lower() == 'antiquus':
        antiq = antiq[i+1]
print('How many bison are of the species antiquus?\nAnswer is:', antiq)
```

### A `mean()` and `std()` functions were called on `bisons.Length` to return the mean and standard deviation for the length

```python
print('What is the mean and standard deviation of Length?\nThe mean is:', bisons.Length.mean(), ', And the standard deviation is:', bisons.Length.std())
```

### And last but not least, the `median()` function was called on `bisons.Length` to return the median length for bisons

```python
print('What is the median length of the bison?\nAnswer is:', bisons.Length.median())
```
